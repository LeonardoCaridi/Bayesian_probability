# Classification

Compute the probability of GRB170817A (T = 2.0 s) of being a short GRB or a long GRB.

Decide a figure of merit and determine the threshold value for T90 to discriminate between short and long GRBs.

```python
import matplotlib.pyplot as plt
import numpy as np

from mcmc_ensemble import thin_chains_by_acf
from mcmc import log_T90_distribution
%load_ext autoreload
%autoreload 2
```

```python
DATA_PATH = "../GRBs/GRB_data.txt"
arr = np.loadtxt(DATA_PATH, skiprows=1)
logT90       = arr[:,0]
sigma_logT90 = arr[:,1]
```

## Parametri medi

```python
def gaussian_function(x, mu, sigma):
    return np.exp(-0.5*((x-mu)/sigma)**2)/(sigma*np.sqrt(2*np.pi))

def gaussian_mixture(x, w, mu1, mu2, sigma1, sigma2):
    return w*gaussian_function(x,mu1,sigma1) + (1-w)*gaussian_function(x,mu2,sigma2)
```

```python
plt.figure(figsize=(8,5))
x = np.linspace(-4,7,10000)
plt.plot(x, gaussian_mixture(x, 0.310, -0.199, 3.548, 1.349, 0.981), linestyle='-.', color='k', label='T90 distribution')
plt.plot(x, 0.310*gaussian_function(x, -0.199, 1.349), label='short GRB')
plt.plot(x, (1-0.310)*gaussian_function(x, 3.548, 0.981), label='long GRB')       

condition = 0.310*gaussian_function(x, -0.199, 1.349) - (1-0.310)*gaussian_function(x, 3.548, 0.981)
thr_index = np.where(condition<=0)[0][0]
thr = x[thr_index]
plt.axvline(thr, linestyle='--', color='r', label=f"threshold={np.exp(thr):.2f} s", linewidth=2)
plt.axvline(np.log(2), linestyle=':', color='grey', label=f"expected thr=2 s", linewidth=2)

plt.hist(logT90, bins=30, density=True, alpha=0.35,
    facecolor='lightgrey', edgecolor='none', label='logT90 histogram', zorder=0)

plt.xlabel('log(T90)')
plt.ylabel('Density')
plt.title('Threshold with mean parameters')
plt.legend()
plt.savefig('../figure/threshold_mean_param.png', dpi=300)
plt.savefig('../figure/threshold_mean_param.pdf', dpi=300)
plt.show()
```

## Modello mediano

Il modello mediano NON corrisponde al modello dei parametri mediani

```python
# Esegue tutti i passagi per ottenere i samples dell'ensamble con dati reali e sigma_logT90
data = np.load("../chains/chains_stack_real_sigmaT90_10.npz")
chains  = data['chains']       # array 3D (n_chains,n_steps,5)

chains_cut = chains.copy()
chains_cut = np.delete(chains_cut, [3,4], axis=0)
print(chains_cut.shape)

burnin = 1000
chains_new = chains_cut[:,burnin:,:]
print(chains_new.shape)

thinned_list, per_param_lags, per_chain_lags = thin_chains_by_acf(
    chains_new, threshold=0.048, max_lag_search=2000
)
for i in range(chains_new.shape[0]):
    print(f'chain {i+1} shape = {thinned_list[i].shape}')

samples = np.concatenate([c for c in thinned_list], axis=0)  # shape (N_total, d)
print(samples.shape)
```

```python
x = np.linspace(-4,7,10000)

# log(T90) distribution
distribution_models = np.vstack([
        np.exp(log_T90_distribution(x, samples[i, :]))
        for i in range(samples.shape[0])
        ]) # shape -> (n_samples, n_points)
    

p_low, p_med, p_high = np.percentile(distribution_models, (5, 50, 95), axis=0)

# Short GRB
distribution_models_short = np.vstack([
        samples[i, 0]*gaussian_function(x, samples[i, 1], np.exp(samples[i, 3]))
        for i in range(samples.shape[0])
        ]) # shape -> (n_samples, n_points)
    
p_low_short, p_med_short, p_high_short = np.percentile(distribution_models_short, (5, 50, 95), axis=0)

# Long GRB
distribution_models_long = np.vstack([
        (1-samples[i, 0])*gaussian_function(x, samples[i, 1] + samples[i,2], np.exp(samples[i, 4]))
        for i in range(samples.shape[0])
        ]) # shape -> (n_samples, n_points)
    
# 2) compute percentiles along the posterior axis (axis=0)
p_low_long, p_med_long, p_high_long = np.percentile(distribution_models_long, (5, 50, 95), axis=0)
```

```python
def threshold_from_arrays(x, y_short, y_long):
    """Restituisce la soglia (x) dove y_short == y_long.
       Usa ricerca del primo cambio di segno e interpolazione lineare.
       Se non c'è crossing, torna il punto di minima differenza."""
    cond = y_short - y_long
    # cerca indice i tale che cond[i] e cond[i+1] hanno segni opposti
    idx = np.where(np.sign(cond[:-1]) * np.sign(cond[1:]) < 0)[0]
    if idx.size > 0:
        i = idx[0]
        # interpolazione lineare per la radice tra x[i] e x[i+1]
        x0, x1 = x[i], x[i+1]
        f0, f1 = cond[i], cond[i+1]
        # evita divisione per zero
        if (f1 - f0) == 0:
            return x0
        xr = x0 - f0 * (x1 - x0) / (f1 - f0)
        return xr
    # fallback: se esiste un punto dove cond <= 0 prendi il primo
    le = np.where(cond <= 0)[0]
    if le.size > 0:
        return x[le[0]]
    # ultimo fallback: minimo valore assoluto della differenza
    return x[np.argmin(np.abs(cond))]

# --- soglia "mediana" (stesso approccio semplice che usavi) ---
thr_med = threshold_from_arrays(x, p_med_short, p_med_long)
print("Threshold (median) = ", thr_med)

# --- metodo conservativo (combinazioni estreme delle bande 5-95%) ---
thr_ext1 = threshold_from_arrays(x, p_low_short,  p_high_long)   # una estremità
thr_ext2 = threshold_from_arrays(x, p_high_short, p_low_long)    # l'altra estremità
thr_min = min(thr_ext1, thr_ext2)
thr_max = max(thr_ext1, thr_ext2)
print("Conservative interval from extremes: [{:.4f}, {:.4f}]".format(thr_min, thr_max))

# --- Monte Carlo semplice usando campionamento uniforme fra le bande ---
nreal = 2000
rng = np.random.default_rng(0)
roots = np.empty(nreal)
for k in range(nreal):
    samp_short = rng.uniform(p_low_short, p_high_short)
    samp_long  = rng.uniform(p_low_long,  p_high_long)
    roots[k] = threshold_from_arrays(x, samp_short, samp_long)

lo5, med50, hi95 = np.percentile(roots, [5, 50, 95])
print("MC 50% = {:.4f}, 90% CI (5-95) = [{:.4f}, {:.4f}]".format(med50, lo5, hi95))
```

```python
fig2, ax2 = plt.subplots(figsize=(8,5))

# background histogram of observations
ax2.hist(logT90, bins=30, density=True, alpha=0.35,
facecolor='lightgrey', edgecolor='none', label='logT90 histogram', zorder=0)

# log(T90) distribution
ax2.fill_between(x, p_low, p_high, alpha=0.4,
facecolor='turquoise', zorder=1)

ax2.plot(x, p_med, 'k-.', label=f'T90 dist median', zorder=2)

# Short GRB
ax2.fill_between(x, p_low_short, p_high_short, alpha=0.4,
facecolor='C0', zorder=1)

ax2.plot(x, p_med_short, 'C0-', label=f'Short GRB', zorder=2)

# Long GRB
ax2.fill_between(x, p_low_long, p_high_long, alpha=0.4,
facecolor='C1', zorder=1)

ax2.plot(x, p_med_long, 'C1-', label=f'Short GRB', zorder=2)

plt.axvspan(lo5, hi95, alpha=0.4, facecolor='C1', zorder=1)
plt.axvline(med50, linestyle='--', color='r', label=f"threshold={np.exp(med50):.2f} s", linewidth=1.5)
plt.axvline(np.log(2), linestyle=':', color='grey', label=f"expected thr=2 s", linewidth=1.5)

plt.xlabel('log(T90)')
plt.ylabel('Density')
plt.title('Threshold with median model')
plt.legend()
plt.savefig('../figure/threshold_median_model.png', dpi=300)
plt.savefig('../figure/threshold_median_model.pdf', dpi=300)
plt.show()
```
